Preparing a Dataset
* How does the dataset handle invalid values?
* What do we want to do with null values?
* Do we want to summarize, group or filter the data?

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Diabetes.csv')
df.info()
```

```python
df.head()
```

```python
df = df.fillna(0)
```

```python
df.head()
```

```python
df2 = df[['Glucose', 'BMI', 'Age', 'Outcome']]
```

```python
df2.head()
```

```python
df2.describe()
```

```python
# remove the row if any of the Glucose/BMI/Age has value 0

# any(axis=1): horizontally, if any of the value has 0
# any(axis=0): vertically, if any of the value has 0
# ~: inverse

df3 = df2.loc[~(df2[df2.columns[:-1]] == 0).any(axis=1)]
df3.describe()
```

```python
df3.info()
```

```python
df3.groupby('Outcome').mean()
```

```python
df3.groupby('Outcome').agg({'Glucose':'mean','BMI':'median','Age':'sum'})
```

```python
df3.groupby('Outcome').agg(['mean','median'])
```

```python
positive = df3.loc[df3['Outcome']==1]
negative = df3.loc[df3['Outcome']==0]
print(positive.shape, negative.shape)
```

```python
df3.to_csv('clean_diabetes.csv', index = False)
```
